```python
from pkasolver.query import mol_query
from pkasolver.query import smiles_query
from pkasolver.query import inchi_query
from pkasolver.query import sdf_query
from pkasolver.query import draw_sdf_mols
from pkasolver.query import draw_pka_map 
from pkasolver.query import draw_pka_reactions
from pkasolver.query import set_model_path 
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from rdkit import Chem
IPythonConsole.molSize = 400,400

output_folder = "output"
# sdf_path = "/data/shared/projects/pkasolver-data-clean/00_AvLiLuMoVe_testdata.sdf"
sdf_path = "../data/Baltruschat/AvLiLuMoVe_cleaned_mono_unique_notraindata.sdf"
```

```python
mols, pkas, atoms = smiles_query("O=C([O-])C[NH+](CCN(CC(=O)[O-])CC(=O)[O-])CC(=O)[O-]", output_smiles=True)
print("Smiles query results (smiles output):")
print(f"mols_list: {mols},\npka_list: {pkas},\natom_list: {atoms} \n")

mols, pkas, atoms = inchi_query(
    'InChI=1S/C10H16N2O8/c13-7(14)3-11(4-8(15)16)1-2-12(5-9(17)18)6-10(19)20/h1-6H2,(H,13,14)(H,15,16)(H,17,18)(H,19,20)/p+2', 
    output_inchi=True)
print("Inchi query results (inchi_output):")
print(f"mols_list: {mols},\npka_list: {pkas},\natom_list: {atoms} \n")

mols, pkas, atoms = smiles_query("O=C([O-])C[NH+](CCN(CC(=O)[O-])CC(=O)[O-])CC(=O)[O-]")
print("Smiles query results (mol output):")
print(f"mols_list: {mols},\npka_list: {pkas},\natom_list: {atoms} \n")
```
